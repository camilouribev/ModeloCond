# Modelo Condensación por gotas en presencia de no condensables 
   

<p style='text-align: justify;'>
Este código implementa algunas fórmulas para calcular la transferencia de calor en la condensación por gotas de gases de pirólsis, teniendo en cuenta la distribución de tamaños de gotas, basado enel modelo de  Kim y Kim (2011) [1], con modificaciones al código desarrollado por Jakob Sablowski disponible 
<a href="https://jsablowski.github.io/blog/DWCmod/"> <strong><i>aquí</i></strong></a>.

En este modelo se hace lo siguiente:

<p> 1. Conociendo una distribución de tamaño de gotas(balance de población), se calcula la transferencia de calor a través de cada gota, teniendo en cuenta un conjunto de resistencias térmicas.&nbsp;</p>

<p>  2. Se halla el porcentaje de área mojada por las gotas y área seca de acuerdo a la configuración del sistema.&nbsp;</p>

<p> 3. Se hace la sumatoria de calores de cada gota individual a través de una integral. Adicionalmente, conociendo el área seca, se le adiciona el calor transferido por convección.&nbsp;</p>


Se debe aclarar que este modelo depende de la tasa de condensación y los puntos de nucleación, para los que actualmente se usan valores hallados en la literatura, pero que son valores experimentales.
 </p>


## Sistema de  Resistencias térmicas
<p>&nbsp;</p>

El modelo calcula el esquema de resistencias térmicas bajo la presencia de no condensables:


Tras plantear el sistema de ecuaciones dependientes del radio, con unos parámetros de tamaño de gota mínima y máxima y una distribución de tamaños se integran el valor de $q$ para todas las gotas y se tiene una medida de la transferencia de calor por unidad de área. Las resistencias son representadas como caídas de temperaturas en cada región de la gota


#### 1. Resistencia asociada a la conducción de no condensables y a la interfase vapor-líquida :
<p>&nbsp;</p>


$$\Delta T_{int}=\frac{q(r)}{2\pi r^2h_i(1-cos\theta)}$$

donde el coeficiente de transferencia de calor interfacial $h_i$ en presencia de no condensables puede expresarse como [3]:

$$h_i = h_{conv}+ \frac{h_{fg}\rho_{mezcla}D Sh}{L_c(T_b-T_i)}ln \frac{1-\omega_{gases cond,i}}{1-\omega_{gases cond,bulk}}$$

Las fracción molar del vapor en la interfase se obtiene mediante la expresión que la relaciona con la temperatura para vapor saturado, publicada por Matuszkiewicz y Vernier[]:

$$\omega(T)= \frac{\frac{M_{cond}P_{sat}(T)}{M_{NC}P}}{1-\frac{(1-\frac{M_{cond}}{M_{NC}})P_{sat}(T)}{P}}$$

#### 2.  Resistencia a la conducción en las gotas
<p>&nbsp;</p>

$$\Delta T_{gota}=\frac{q_d\theta}{4\pi rk_{condensado} sin\theta}$$

#### 3. Resistencia asociada al recubrimiento hidrofóbico
<p>&nbsp;</p>

$$\Delta T_{coat}=\frac{q_d\delta}{4\pi r²k_{coat} sin^2\theta}$$

#### 4. Resistencia por curvatura de la gota
<p>&nbsp;</p>
dado que el radio mínimo de gota termodinámicamente viable está dado por la diferencia de temperatura entre el vapor y la superficie, tenemos que:

$$r_{min}=\frac{2T_{sat}\sigma}{h_{fg}\rho\Delta T}$$
<p>&nbsp;</p>
Por lo que podemos escribir la resistencia asociada a la curvatura de la gota como:

$$\Delta T_{curv}=\frac{r_{min}}{r\rho}\Delta T$$

<p>&nbsp;</p>

Tenemos que la diferencia total de temperatura entre los gases y la superficie está dada por la suma de los delta de T:

$$\Delta T = \Delta T_{int} + \Delta T_{gota} + \Delta T_{coat}+ \Delta T_{curv} $$
<p>&nbsp;</p>
Reorganizando los términos, tenemos la expresión para el flux de calor para cada gota, dependiente del radio r:
<p>&nbsp;</p>
$$q_d=\frac{\Delta T \pi r²(1-r_{min}/r)}{\frac{\delta}{k{coat}sin²\theta}+\frac{r\theta}{4k_c sin \theta}+\frac{1}{2h_i(1-cos_{\theta})}}$$
<p>&nbsp;</p>

### Distribución de gotas

Para la implementación del modelo, es necesario conocer la distribución de gotas y los radios mínimo y máximo. Para la distribución de gotas pequeñas, se usó la distribución establecida por LeFevre y Rose en el artículo _“A Theory of Heat Transfer by Dropwise
Condensation”_ dada por :
<p>&nbsp;</p>
$$ N(r)=\frac{1}{3\pi r^2r_{max}}\left(\frac{r}{r_{max}}\right)^{(-2/3)}$$

<p>&nbsp;</p>
Para las gotas pequeñas, se usó la distribución propuesta por Kim y Kim en  _“Dropwise condensation modeling suitable for superhydrophobic surfaces”_:
<p>&nbsp;</p>
$$ n(r)=\frac{1}{3\pi r^3r_{max}}\left(\frac{r}{r_{max}}\right)^{(-2/3)}\frac{r(r_e-r_{min})A_2r+A_3}{r-r_{min})A_2r_e+A_3}exp(B_1+B_2)$$
<p>&nbsp;</p>
Donde $A_2,A_3,B_1,B_2$ son valores dependientes del ángulo de contacto y las temperaturas de la corriente de gases y la superficie 

### Radio de gotas

Se calculan los radios mínimos, máximo y efectivo(radio que separa las dos categorías), a partir  del trabajo de Kim y Kim. Estos radios dependen de la tensión superficial del líquido y el ángulo de contacto que haga con la superficie:

$$ r_{min}=\frac{2T_{sat}\sigma}{H_{fg}\rho\Delta T}$$
<p>&nbsp;</p>

$$r_e=(4N_s)^{-1/2}$$
<p>&nbsp;</p>

$$r_{max}=\left(\frac{6c(cos\theta_r-cos\theta_a)sin\theta \sigma}{\pi(2-3cos\theta+cos^3\theta)\rho g}\right)$$



### Flux de calor

Finalmente, el flux de calor por unidad de área está dado por la suma de integrales
<p>&nbsp;</p>

$$q^{"}=\int_{r_{min}}^{r_e}q_d(r)n(r)dr+\int_{r_{e}}^{max}q_d(r)N(r)dr+ h_{conv}f_{seca}\Delta T$$
<p>&nbsp;</p>

```python
## Importar librerías 

import math
import numpy as np
import scipy
import scipy.integrate as integrate
import sys
from functools import partial
from thermo.chemical import Mixture, Chemical
import statistics
from IPython.display import Image
```

### Parámetros de entrada

```python
T_g=399 #TEMPERATURA DE GASES
T_s=273.15 #TEMPERATURA DE SUPERFICIE
T_propiedades= T_s+(0.25*(273.25-T_s))


P=101325 # Presión de operación [Pa]

Frac_n2=0.567 #fracción másica de nitrógeno en los gases de pirólisis
Frac_vap = 1-Frac_n2
frac_w = (1-Frac_n2)*0.2 # Se asume que el 50% delos gases es bioaceite y que 20% del bioaceite es agua.
frac_comp=(1-Frac_n2-frac_w)/7 # Fracción másica de los compuestos en los gases de pirólisis (teniendo en cuenta el N2)
frac_non=(1-frac_w)/6 # Fracción másica de los compuestos en el condensado (se asumen partes iguales)

t_c = (2*0.5*0.8)/(3600*1000)            #Tasa de condensación requerida en el condensador [kg/s]


alfa=0.02    #Factor de acomodación Wikramanayake et al
theta_degrees=65     #Ángulo de contacto en grados

delta_coat=2e-06 #2e-06 CYTOP
k_coat=0.12 # 0.12 CYTOP

N_s= 1*10**9 # Tomado de Leach et al - Numero de puntos de nucleación 
HAC=10 #Histéresis de ángulo de contacto
```

```python
#Composición de los gases de pirólisis
comp = {'Nitrogen': Frac_n2,
        'water': frac_w,
        'acetic acid': frac_comp,
        'ethanol': frac_comp,
        'acetaldehyde': frac_comp,
        'propionic acid': frac_comp,
        'butanone': frac_comp,
        'acetol':frac_comp }

#Composición del condensado
compcond = {'water': frac_w,
            'acetic acid': frac_non,
            'ethanol': frac_non,
            'acetaldehyde': frac_non,
            'propionic acid': frac_non,
            'butanone': frac_non,
            'acetol': frac_non}


condensado= Mixture(ws=compcond, T=T_propiedades, P=P)


T_sat =min([x for x in condensado.Tbs if x is not None])# Temperatura de saturación K
T_sal=T_sat-10 # Temperatura de salida [K]
T_prom = (T_g+T_sal)/2 # Temperatura promedio [K]
T_i=T_sat #Temperatura de saturación [K]


vap_sat= Mixture(ws=compcond, T=T_i, P=P)
gases = Mixture(ws=comp, T=T_prom, P=P)#ws: fracción másica, zs: Fracción molar, Vfls: Fracción volumétrica líquida, Vfgs: Fracción volumétrica gaseosa


rho_gases= gases.rhog # densidad de la mezcla gaseosa [Kg/m³]
rho= statistics.mean([x for x in condensado.rhols if x is not None]) # densidad de la mezcla líquida [kg/m³]
k_gas= gases.kg  #conductividad térmica de los gases [W/m K]
k_c= condensado.kl # Conductividad térmica del condensado [W/mK]
cp_cond=condensado.Cpl # Calor específico del condensado [ J/(kg⋅K)]
h_fg=statistics.mean([x for x in condensado.Hvaps if x is not None]) # Hvaps es entalpía de evaporación en [J/kg]
v_g=(1/rho_gases) # Volúmen específico del gas [m3/kg]
sigma=statistics.mean([x for x in condensado.sigmas if x is not None]) # tensión superficial [N/m]

P_sat=statistics.mean([x for x in condensado.Psats if x is not None]) # presion de saturación de los gases en [Pa]
P_sat_i=statistics.mean([x for x in vap_sat.Psats if x is not None]) # presion de saturación del vapor en la interfase [Pa]

C_con=gases.rhogm # densidad molar de la mezcla [mol/m3]

M_n2= Chemical('Nitrogen', T=(T_prom)).rhog/Chemical('Nitrogen', T=(T_prom)).rhogm #masa molar N2 [kg/mol]
M_vap = Mixture(ws=compcond, T=T_prom, P=P).rhog/Mixture(ws=compcond, T=T_prom, P=P).rhogm # masa molar de vapores condensables [kg/mol]

R_g=8.314/(rho_gases/C_con) # Constante específica del gas [J/kgK] 

mm=M_vap/M_n2 # Relación de masas molares condensables y no condensables
Frac_vap_i = (mm*P_sat_i/P)/(1-(1-mm)*P_sat_i/P) # Fracción de vapor en la interfase [18] Matuszkiewicz, A., & Vernier, P. (1991). 
                                                #Two-phase structure of the condensation boundary layer with a non-condensing gas and liquid droplets. 

    
#h_i=((2*alfa)/(2-alfa))*(1/math.sqrt(2*math.pi*R_g*T_s))*((h_fg**2)/(v_g*T_s))

h_conv_film= 3.58 #c Coeficiente de convección [W/m²K]

D_global=5.169735467575174e-06 # Coeficiente de difusión global
Sh=14.9642 # Número de Sherwood
Lc=0.07792 # Longitud característica
delta= Lc/Sh#  Espesor de la capa límite, Tomado del número de sherwood por analogía de transferencia de masa [peterson 1993][m] 


h_i = h_conv_film + h_fg*rho_gases*D_global*Sh*np.log((1-Frac_vap_i)/(1-Frac_vap))/Lc*(delta*(T_prom-T_i)) # Coeficiente interfacial de transferencia de calor [W/m2K]


h_i
```

```python
#Límites de integración

#Radio mínimo

def r_min_KimKim(T_sat, sigma, h_fg, rho, T_s):

    r_min = 2*T_s*sigma / (h_fg * rho * (T_sat-T_s)) 	
    return r_min

theta_r=theta_degrees-(0.5*HAC)
theta_a=theta_degrees+(0.5*HAC)


theta, theta_a, theta_r = [math.radians(x) for x in [theta_degrees,theta_a, theta_r]]


const=1 #constante numérica depende de la forma de la gota y la inclinación de la superficie
g=9.81


#Radio máximo

def r_max_KimKim(const, theta_r, theta_a, theta, sigma, rho, g):

    r_max = math.sqrt((6*const*(math.cos(theta_r)-math.cos(theta_a))*math.sin(theta)*sigma)/ \
                      (math.pi*(2-3*math.cos(theta)+(math.cos(theta)**3))*rho*g))
    
    return r_max

r_min=r_min_KimKim(T_sat=T_sat,sigma=sigma,h_fg=h_fg,rho=rho,T_s=T_s)
r_max=r_max_KimKim(const=const,theta_r=theta_r,theta_a=theta_a,theta=theta,sigma=sigma,rho=rho,g=g)

#N_s=(0.037/((r_min)**2)) # sitios de nucleación 1/m²

print("delta:",delta,"\n"
    "r_min:",r_min,"m\n"
     "r_max:",r_max,"m\n"
     "Theta:",theta,"\n"
    'N_s:{:.2E}'.format(N_s),'Sitios de nucleación [1/m²]')


```

```python
deltaT_sub=T_prom-T_s


def A_mojada(r):
    A_m= math.pi*(r**2)*(math.sin(theta))**2
    return A_m



def N_LeFevre(r, r_max):
    "Distribución de tamaño de gotas para gotas grandes "
    N = 1/(3*math.pi*r**2*r_max) * (r/r_max)**(-2/3)
    return N

    
def r_e_KimKim(N_s):
    """ Radio efectivo de la gota """
    r_e = (4*N_s)**(-0.5)
    return r_e


r_e=r_e_KimKim(N_s)
    
def n_KimKim(r, deltaT_sub, r_min, delta_coat, k_coat, k_c, theta, h_i, rho, h_fg, r_e, r_max):
    """ Distribución de tamaño de gotas para gotas pequeñas """
    A_1 = deltaT_sub / (2*rho*h_fg)
    A_2 = theta * (1-math.cos(theta)) / (4*k_c*math.sin(theta))
    A_3 = (1/(2*h_i)) + delta_coat*(1-math.cos(theta)) / (k_coat*(math.sin(theta))**2)
    tau = 3*r_e**2 * (A_2*r_e + A_3)**2 / (A_1*(11*A_2*r_e**2 - 14*A_2*r_e*r_min + 8*A_3*r_e - 11*A_3*r_min))
    B_1 = A_2/(tau*A_1) * ((r_e**2-r**2)/2 + r_min*(r_e-r) - r_min**2*math.log((r-r_min)/(r_e-r_min))) 
    B_2 = A_3/(tau*A_1) * (r_e-r - r_min*math.log((r-r_min)/(r_e-r_min)))
    n = 1/(3*math.pi*r_e**3*r_max) * (r_e/r_max)**(-2/3) * r*(r_e-r_min)/(r-r_min) * (A_2*r+A_3)/(A_2*r_e+A_3) * math.exp(B_1+B_2)
    return n 


def Q_got_KimKim(r,t_c, h_fg, h_i, sigma, k_gas, delta, theta, alfa, T_s,k_c, T_prom,deltaT_sub, k_coat, r_min):
    "Resistencia total por gotas pequeñas"""
    #T_int=T_prom-sigma/(h_i*2*math.pi*(r**2))
    
    R_dif= (T_prom-T_s)/(t_c*h_fg)
    R_cond=delta/(k_gas*math.pi*(r**2)*math.sin(theta)**2)
    R_int=1/(h_i*2*math.pi*(r**2)*(1-math.cos(theta)))
    R_drop=(theta/(4*math.pi*k_c*r*math.sin(theta)))
    R_tot = (((1/R_dif)+(1/R_cond))**-1)+R_int+R_drop
    
    #Q_got=(T_prom-T_s)/((R_dif*R_cond/(R_dif+R_cond))+R_int+R_drop)
    #Q_got=(T_prom-T_s)/R_tot
    
    Q_got = (deltaT_sub * math.pi * r**2 * (1 - r_min/r)) / \
        (delta_coat/(k_coat * (math.sin(theta))**2) + r*theta/(4*k_c*math.sin(theta)) + 1/(2*h_i*(1-math.cos(theta))))
    return Q_got




Q_got = partial(Q_got_KimKim, t_c=t_c, h_fg=h_fg, h_i=h_i, sigma=sigma, k_gas=k_gas, delta=delta, theta=theta,
                 alfa=alfa, T_s=T_s, k_c=k_c, T_prom=T_prom, deltaT_sub= deltaT_sub, k_coat=k_coat, r_min=r_min)


    
A_m=partial(A_mojada)
A_m.__doc___="Área mojada por la gota"

n = partial(n_KimKim, deltaT_sub=deltaT_sub, r_min=r_min, delta_coat=delta_coat, k_coat=k_coat, k_c=k_c,
                theta=theta, h_i=h_i, rho=rho, h_fg=h_fg, r_e=r_e, r_max=r_max)
n.__doc__ = " Distribución de tamaño de gotas para tamaños pequeños dependiendo del radio r en m"
N = partial(N_LeFevre, r_max=r_max)
N.__doc__ = "Distribución de tamaño de gotas para gotas grandes, r en m"


def Q_drop_n(r):
    Q_drop_n = Q_got(r)*n(r)
    return Q_drop_n


def Q_drop_N(r):
    Q_drop_N = Q_got(r) * N(r)
    return Q_drop_N



def A_moj_n(r):
    A_moj_n= A_m(r)*n(r)
    return A_moj_n


def A_moj_N(r):
    A_moj_N = A_m(r) * N(r)
    return A_moj_N

"""Integración del flujo de calor de acuerdo a la distribución de gotas"""

A_n,A_n_interr= integrate.quad(A_moj_n,r_min,r_e)
A_N,A_N_interr= integrate.quad(A_moj_N,r_e,r_max)
                                   
q_n, q_n_interr = integrate.quad(Q_drop_n, r_min, r_e )
q_N, q_N_interr = integrate.quad(Q_drop_N, r_e, r_max)

AREA_MOJ= A_n+A_N


q_conv=(1-AREA_MOJ)*h_conv_film*deltaT_sub # Calor asociado a la parte seca [W/m2] 

q = q_n + q_N+ q_conv #[W/m2]


print ("h_i:", round(h_i,2), "W/m2k\n"
      "r_e:",r_e,"\n"
      "q_N:",q_N,"W/m2\n"
      "q_n:",q_n,"W/m2\n"
       "q_conv:",q_conv,"W/m2\n"
      "q:", round(q,2),"W/m2\n"
      "Relacion q_N/q:",round((q_N*100/(q_N+q_n)),2),"%\n"
       "T salida: {} K".format(T_sal) 
       )
```

### Nomenclatura&nbsp;

$R_{dif}$:resistencia térmica por la capa de difusión

$T_{amb}$: Temperatura gases de pirólisis [K]

$T_{int}$: Temperatura de la interfase de la gota [K]

$t_c$: Tasa de condensación por área. Hallada experimentalmente [Kg/m²s]

$h_{fg}$: Calor latente de condensación [J/kg]

$k_{nc}$: Conductividad térmica de la mezcla de no condensables [W/m K]

$\delta$: grosor de la capa límite [m]

$S_r$: Área proyectada de la gota [m²]

$R_g$: Constante específica de los gases (8.3144598 / masa molar)    J/[Kg K]

$v_g$: Volumen especifico del vapor de pirólisis (1/densidad vapor) [m²/Kg]

$h_{fg}$: Calor latente de vaporización [J/Kg]

$\alpha$: Coeficiente de acomodación o condensación; fracción de moléculas de vapor que se mueven en la gota líquida durante el cambio de fase. Debe ser de entre 0.02 a 0.04 para agua. En Wrikamanayake et al. se usó 0.02

$T_s$: Temperatura de superficie de pared [K]

$T_{sat}$: Temperatura de saturación del vapor [K]

$k_1$: Conductividad térmica del condensado [W/mk]


$r$: radio de la gota [m]

$\theta$: Ángulo de contacto [°]

$\delta_{coat}$: Espesor de la capa recubrimiento hidrofóbica [m]

$k_{coat}$: Conductividad térmica del recubrimiento hidrofóbico [W/mK]

$\rho$: Densidad del condensado [Kg/m³]

$k_c$: Conductividad térmica del condensado [W/mK]

$\sigma$: Tensión superficial [N/m]

$N_s$: Número de sitios de nucleación en 10⁹ 1/m² 

$r_{min}$: Radio de gota mínimo [m]

$r_{max}$: Radio de gota máximo [m]

$c$: Concentración vapor de pirólisis en los gases [mol/m³]
